####Sequence Slicing

Method to get slice of elements from list/string.

####List comprehensions: Create new list based on existing list

```python
# y = x + 1

x_list = [0, 1, 2, 3, 4]


y_list = []
for x in x_list:
    y_list.append(x+1)
print(y_list)
```

```python
# List comprehension

x_list = [0, 1, 2, 3, 4]
y_list = [x+1 for x in x_list]
print(y_list)
```

```python
# List comprehension with conditionals
# y = x+1 if x%2==0 i.e. if x is even

x_list = [0, 1, 2, 3, 4]
y_list = [x+1 for x in x_list if x%2==0]
print(y_list)
```

###Functions

```python
def isEven(x, strReturn=False):
  ''' Return True if x is even else False
  '''
  isEven = x%2 == 0
  if strReturn:
    if isEven:
      retVal = 'TRUE'
    else:
      retVal = 'FALSE'
  else:
    retVal = isEven
  return retVal 

x_list = [0, 1, 2, 3, 4]
for x in x_list:
    print(x, isEven(x))
```

```python
x_list = [0, 1, 2, 3, 4]
a=[1,2,3]
b=[1,2,3]
c=a+b
print(c)
for x in x_list:
    print(x, isEven(x, strReturn=True))
```

```python
# 1. Row based addition I
import numpy as np
a = np.array([[1,2,3], [4,5,6]])
b = np.array([0,0,1])
c = np.zeros(a.shape)
# Write a loop to add [b] to each row of [a] and assign to new array [c].
'''
 c =  [[1,2,3]  +   [[0,0,1]    = [[1,2,4],
       [4,5,6]]      [0,0,1]]      [4,5,7]]
'''
# START

d=b
b=np.concatenate((d,d))
b=b.reshape(2,-1)

for i in range(len(a)):
  for j in range(len(a[0])):
    c[i][j] = a[i][j] + b[i][j]
print(c)
# END

print((c==np.array([[1,2,4], [4,5,7]])).all())
```

```python
# 2. Row based addition II. 
# Wihout a loop, add [b] to each row of [a] and assign to new array [c], by using [tile: https://numpy.org/doc/stable/reference/generated/numpy.tile.html] function
'''
 c =  [[1,2,3]  +   [[0,0,1]    = [[1,2,4],
       [4,5,6]]      [0,0,1]]      [4,5,7]]
'''
import numpy as np
a = np.array([[1,2,3], [4,5,6]])
b = np.array([0,0,1])
c = np.zeros(a.shape)

# START
b = np.tile(b,reps=(2,1))
c= a + b
print(c)
# END

print((c==np.array([[1,2,4], [4,5,7]])).all())
```
